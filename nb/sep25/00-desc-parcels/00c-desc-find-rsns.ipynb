```python
import pandas as pd

# 1. This is the new, correct URL for the raw CSV file
new_url = 'https://raw.githubusercontent.com/int-brain-lab/ibllib/main/ibllib/atlas/allen_structure_tree.csv'

# 2. Load the file directly into a pandas DataFrame
try:
    allen_map_df = pd.read_csv(new_url)

    # 3. This is your list of Allen-derived ROI acronyms (replace with your actual list)
    my_rois = ['MOp5', 'SSp-bfd4', 'VISam4', 'PL2', 'AUDd1', 'ORBl5']
    my_rois_df = pd.DataFrame(my_rois, columns=['acronym'])

    # 4. Merge your ROIs with the mapping dataframe to assign RSNs
    grouped_rois_df = pd.merge(my_rois_df, allen_map_df[['acronym', 'rsn_name']], on='acronym', how='left')

    # 5. Display the final result
    print("Your ROIs mapped to RSNs:")
    print(grouped_rois_df)

except Exception as e:
    print(f"An error occurred: {e}")
    print("Please ensure you have an internet connection and the pandas library is installed (`pip install pandas`).")
```
