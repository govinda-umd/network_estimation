```python
import csv
import os
import sys
import numpy as np
import pandas as pd
import scipy as sp 
import pickle 
import json
from os.path import join as pjoin
from itertools import product, permutations
from tqdm import tqdm as tqdm
from copy import deepcopy

import glob

# plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.cm import rainbow

plt.rcParamsDefault['font.family'] = "sans-serif"
plt.rcParamsDefault['font.sans-serif'] = "Arial"
plt.rcParams['font.size'] = 14
plt.rcParams["errorbar.capsize"] = 0.5

import cmasher as cmr  # CITE ITS PAPER IN YOUR MANUSCRIPT

# ignore user warnings
import warnings
warnings.filterwarnings("ignore") #, category=UserWarning)
```

```python
class ARGS():
    pass

args = ARGS()

args.SEED = 100
```

```python
args.num_rois = 58
args.roi_size = 600
args.maintain_symmetry = True
args.brain_div = 'whl'

DESC = (
    f'type-functional'
    f'_nrois-{args.num_rois}'
    f'_size-{args.roi_size}'
    f'_symm-{args.maintain_symmetry}'
    f'_braindiv-{args.brain_div}'
)

ELBOS = pd.DataFrame()
BASE_path = f'{os.environ["HOME"]}/mouse_dataset/roi'
for args.num_comms in tqdm(np.arange(3, 12)):
    COMM_path = f'{BASE_path}/{DESC}/svinets/k-{args.num_comms}'
    files = glob.glob(f'{COMM_path}/**/max.txt', recursive=True)
    elbos = [np.loadtxt(file)[2] for file in files]
    ELBOS = pd.concat([
        ELBOS, 
        pd.DataFrame({
            'k':[args.num_comms,]*len(elbos),
            'elbo':elbos,
        }),
    ])
ELBOS = pd.DataFrame(ELBOS)

fig, axs = plt.subplots(1, 1, figsize=(11, 5), dpi=90)
ax = axs
sns.stripplot(data=ELBOS, x='k', y='elbo', ax=ax, alpha=0.1, size=1)
sns.pointplot(data=ELBOS, x='k', y='elbo', ax=ax, color='orange', errorbar='sd', capsize=.4, markersize=10)
ax.grid(True)
ax.set(ylim=[-1.1, 0.0], xlabel=f'#comms.', ylabel=f'elbo value', title=f'{DESC}')
```
