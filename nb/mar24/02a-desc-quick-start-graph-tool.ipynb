```python
import os
import numpy as np
import graph_tool.all as gt
import pickle

# plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# ignore user warnings
import warnings
warnings.filterwarnings("ignore") #, category=UserWarning)
```

```python
class ARGS():
    pass

args = ARGS()

args.SEED = 100
args.num_repeats = 100

gt.seed_rng(args.SEED)
np.random.seed(args.SEED)
```

```python
SIM_path = f'{os.environ["HOME"]}/mouse_dataset/simulations/02a'
```

```python
# RUN THIS SCRIPT BEFORE THIS .IPYNB
# by giving all the parameters
# python 02a-desc-simulated-graphs-fit-sbm.py -bs 40 60 -bd 0.3 0.3 -os 6 -od 0.3 -sbm o -dc False -w 1000 -ni 10 -s 100 &

# the .npy file from above is expected before running the following cells in this notebook.
```

---

```python
args.block_sizes = [40, 60]
args.block_densities = [0.3, 0.3]
args.overlap_sizes = [14]
args.overlap_densities = [0.3]
```

```python
args.sbm_state = 'd'
args.deg_corr = True

file_name = (
    f'bs-{args.block_sizes}'
    f'_bd-{args.block_densities}'
    f'_os-{args.overlap_sizes}'
    f'_od-{args.overlap_densities}'
    f'_sbm-{args.sbm_state}'
    f'_dc-{args.deg_corr}'
    f'.npy'
)

with open(f'{SIM_path}/{file_name}', 'rb') as f:
    [marginal, pmode, g, state, bs, Bs, mdls] = pickle.load(f)
    

nrows, ncols = 1, 2
fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=(5*ncols, 4*nrows), dpi=90)
fig.tight_layout()

ax = axs[0]
sns.heatmap(gt.adjacency(g).todense(), ax=ax)

ax = axs[1]
gt.graph_draw(g, vertex_text=g.vertex_index, ax=ax)


ncols = 1
nrows = 1
fig, axs = plt.subplots(
    nrows=nrows, ncols=ncols, 
    figsize=(4*ncols, 6*nrows), dpi=90,
    sharex=True, sharey=True
)
fig.tight_layout()
ax = axs
sns.heatmap(marginal, ax=ax)
ax.set(
    title=f'marginals',
    xlabel='block', ylabel='roi',
    xticklabels=np.arange(marginal.shape[-1])+1,
)


nrows, ncols = 1, 2
fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=(5*ncols, 4*nrows), dpi=90)
fig.tight_layout()

ax = axs[0]
# sns.histplot(mdls, stat='proportion', ax=ax)
ax.hist(mdls)
ax.set(title=f'MDL', xlabel='MDL', ylabel='proportion')

ax = axs[1]
sns.barplot(Bs / np.sum(Bs))
ax.set(title=f'num. blocks', xlabel='B', ylabel=f'$P(B | A)$', xlim=[1, 12])
```
